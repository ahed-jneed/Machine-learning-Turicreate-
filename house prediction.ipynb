# launch turicreate


```python
import turicreate
import  as agg
```

# Load house sales data

```python
sales = turicreate.SFrame("home_data.csv")
```

```python
sales
```

# Explor

```python
sales.show()
```

```python
turicreate.show(sales[1:5000]['sqft_living'],sales[1:5000]['price'])
```

# Simple regression model that predicts price 

```python
training_set,test_set = sales.random_split(.8,seed=0)
```

# Train simple regression model

```python
model1 = turicreate.linear_regression.create(training_set,target = 'price')
```

# Evaluate the quality of our model

```python
sales['price'].mean()
```

```python
sales['price'].max()
```

```python
print(model1.evaluate(test_set))
```

# Explore model a little further

```python
model1.coefficients
```

# Explor other features of the data

```python
advanced_features = [
'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'zipcode',
'condition', # condition of house				
'grade', # measure of quality of construction				
'waterfront', # waterfront property				
'view', # type of view				
'sqft_above', # square feet above ground				
'sqft_basement', # square feet in basement				
'yr_built', # the year built				
'yr_renovated', # the year renovated				
'lat', 'long', # the lat-long of the parcel				
'sqft_living15', # average sq.ft. of 15 nearest neighbors 				
'sqft_lot15', # average lot size of 15 nearest neighbors 
]
```

```python
sales[advanced_features].show()
```

# Explor zipcode data with price


```python
turicreate.show(sales[1:5000]['zipcode'],sales[1:5000]['price'])
```

# Compute the number of occurrences of each zipcode

```python
zip_count = sales.groupby(key_column_names = 'zipcode',
                         operations={'COUNT':agg.COUNT()})
zip_count
```

```python
zipcode_avg_price = sales.groupby(key_column_names = 'zipcode',
                                  operations={'mean_price' : agg.MEAN('price')}
                                 )
zipcode_avg_price.tail()                   
```

```python
zipcode_avg_price.sort('mean_price',ascending = False)
```

# Bild a model with this additional features


```python
features_model = turicreate.linear_regression.create(training_set,target = 'price',features=advanced_features)
```

# Evaluate quality of our  models with more complex one

```python
print(model1.evaluate(test_set))
print(features_model.evaluate(test_set))
```
