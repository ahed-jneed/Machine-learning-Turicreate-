# Launch turicreate

```python
import turicreate
```

# Read products review data 

```python
products = turicreate.SFrame("amazon_baby.csv")
```

```python
products
```

# Build sentiment classfier


## Build word count vector

```python
products['word_count'] = turicreate.text_analytics.count_words(products['review'])
products
```

# Define the sentiment words To determine if the reviews positive or negative

```python
selected_words = ['awesome', 'great', 'fantastic', 'amazing', 'love', 'horrible', 'bad', 'terrible', 'awful', 'wow', 'hate']
```

```python
for words in selected_words:
    products[words] = products['word_count'].apply(lambda counts: counts.get(words,0))
```

```python
products.tail()
```

```python
#ignore 3-stars review
products = products[products['rating'] != 3]
```

```python
#positive sentiment 4-stars or 5-stars
products['sentiment'] = products['rating'] >= 4
products
```

```python
products[selected_words].show()
```

# Explor how many times words used

```python
for col in selected_words:
    print (col,products[col].sum())
```

# Train our sentiment classifier

```python
training_data,test_data = products.random_split(.8,seed=0)
```

```python
word_count_model = turicreate.logistic_classifier.create(training_data,target='sentiment',features=['word_count'],validation_set=test_data)
```

```python
selected_words_model  = turicreate.logistic_classifier.create(training_data,target='sentiment',features=selected_words,validation_set=test_data)
```

```python
selected_words_model.coefficients
```

```python
selected_words_model.coefficients.sort('value',ascending=False)
```

# Evaluate the quality of our models

```python
print (selected_words_model.evaluate(test_data))
```

```python
print (word_count_model.evaluate(test_data))
```
